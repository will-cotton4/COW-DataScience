```
import pandas as pd
import category_encoders as ce
from sklearn.preprocessing import StandardScaler

dyadic_war_url = 'https://raw.githubusercontent.com/will-cotton4/COW-DataScience/master/directed_dyadic_war.csv'
national_trade_url = 'https://raw.githubusercontent.com/will-cotton4/COW-DataScience/master/National_COW_4.0.csv'
dyadic_trade_url = 'https://raw.githubusercontent.com/will-cotton4/COW-DataScience/master/Dyadic_COW_4.0.csv'
cc_url = 'https://raw.githubusercontent.com/will-cotton4/COW-DataScience/master/COW_country_codes.csv'

dyad_war = pd.read_csv(dyadic_war_url)
cc = pd.read_csv(cc_url)

def clean_dyad_war(df):
  """
  This function cleans the dyadic war dataset:
  - Removes irrelevant/administrative columns
  - Replace country codes with name of country and abbreviation
  - Clean dates and convert to datetime
  - Encode warolea, waroleb, wardyadrolea, wardyadroleb, outcomea (one-hot)
  - Standardize batdtha, batdthb, batdths (commented out)
  - Add Middle East/sub-saharan Africa categorical
  - TODO: (STRETCH) Add count of wars from last ten years
  - TODO: (STRETCH) Add interaction feature between count of wars and warole
  """
  
  df = df.copy()
  
  df = df.drop(columns=['changes_1', 'changes_2', 'disno', 'dyindex'])
  
  # Create dicts to hold CCode:Name and CCode: Abbrv

  name_dict = {row[1]['CCode']: row[1]['StateNme'] for row in cc.iterrows()}
  abbr_dict = {row[1]['CCode']: row[1]['StateAbb'] for row in cc.iterrows()}
  
  df['StateAname'] = df['statea'].replace(name_dict)
  df['StateBname'] = df['stateb'].replace(name_dict)
  df['StateAabbr'] = df['statea'].replace(abbr_dict)
  df['StateBabbr'] = df['stateb'].replace(abbr_dict)
  
  # Repair broken cells and convert to DT
  df['warendyr'] = df['warendyr'].replace({19118:1918})
  df['start_dt'] = pd.to_datetime((df['warstrtmnth'].astype(str) +'/'+ df['warstrtday'].astype(str)+ '/' + df['warstrtyr'].astype(str)).astype(str))
  df['end_dt'] = pd.to_datetime((df['warendmnth'].astype(str) +'/'+ df['warenday'].astype(str)+ '/' + df['warendyr'].astype(str)).astype(str))
  old_dates = ['warstrtmnth', 'warstrtday', 'warstrtyr', 'warendmnth', 'warenday', 'warendyr']
  df = df.drop(columns=old_dates)
  
  # One-hot encoding:
  to_one_hot = ['warolea', 'waroleb', 'wardyadrolea',
                'wardyadroleb', 'outcomea']
  one_hot = ce.OneHotEncoder(cols = to_one_hot)
  df = df.merge(one_hot.fit_transform(df))
  df = df.drop(columns=to_one_hot)
  
  # Standardize:
  # to_scale = ['batdtha', 'batdthb', 'batdths']
  # scaler = StandardScaler()
  # df[to_scale] = scaler.fit_transform(df[to_scale])
  
  # Add categoricals to designate whether in the Middle East or 
  # sub-Saharan Africa
  
  middle_east = ['Bahrain', 'Cyprus', 'Egypt', 'Iran', 'Iraq', 'Israel', 
                 'Jordan', 'Kuwait', 'Lebanon', 'Oman', 'Qatar', 'Saudi Arabia',
                 'Syria', 'Turkey', 'United Arab Emirates', 'Yemen Arab Republic']
  ssa = ['Angola', 'Benin', 'Botswana', 'Burkina Faso', 'Burundi', 'Cameroon',
         'Cape Verde', 'Central African Republic', 'Chad', 'Comoros', 'Congo', 
         'Congo DR', "Côte d'Ivoire", 'Djibouti', 'Equatorial Guinea', 
         'Eritrea', 'Ethiopia', 'Gabon', 'The Gambia', 'Ghana', 'Guinea',
         'Guinea-Bissau', 'Kenya', 'Lesotho', 'Liberia', 'Madagascar', 'Malawi',
         'Mali', 'Mauritania', 'Mauritius', 'Mozambique', 'Namibia', 'Niger', 
         'Nigeria', 'Réunion', 'Rwanda', 'Sao Tome and Principe', 'Senegal',
         'Seychelles', 'Sierra Leone', 'Somalia', 'South Africa', 'Sudan',
         'Swaziland', 'Tanzania', 'Togo', 'Uganda', 'Western Sahara', 'Sahara',
         'Zambia', 'Zimbabwe']
  me_ssa = middle_east + ssa
  
  df['in_me_ssa'] = df['StateAname'].isin(me_ssa)
  return df
dyad_war = clean_dyad_war(dyad_war)
dyad_war
```

```
import pandas as pd

dyadic_war_url = 'https://raw.githubusercontent.com/will-cotton4/COW-DataScience/master/directed_dyadic_war.csv'
national_trade_url = 'https://raw.githubusercontent.com/will-cotton4/COW-DataScience/master/National_COW_4.0.csv'
dyadic_trade_url = 'https://raw.githubusercontent.com/will-cotton4/COW-DataScience/master/Dyadic_COW_4.0.csv'
cc_url = 'https://raw.githubusercontent.com/will-cotton4/COW-DataScience/master/COW_country_codes.csv'

natl_trade = pd.read_csv(national_trade_url)
cc = pd.read_csv(cc_url)

natl_trade = pd.read_csv(national_trade_url)

def clean_natl_trade(df):
  # This method cleans the national trade dataset:
  # - Removes alt_imports and alt_exports; these were specially designated for
  #   Belgium and Luxembourg (who have a special economic relationsip) and China
  #   (due to the special status of Macao and Hong Kong)
  # - Removes version, source1, and source2 (reporting variables, unrelated to 
  #   status of country)
  # - Use forward fill for each country missing import/export data, then back
  #   fill to capture missing data. The remaining rows are dropped, because
  #   they correspond to countries with no import/export data at all.
  # - Standardize imports, exports (commented out)
  # - Add trade deficit feature
  # - Add increase/decrease in imports/exports from past year/past 5 years
  
  df = df.copy()
  
  df = df.drop(columns=['alt_imports', 'alt_exports', 
                        'source1', 'source2', 'version'])
  inc_cond = df['imports'].isnull()
  incomplete = df[inc_cond]['statename'].value_counts().keys()
  
  for state in incomplete:
    state_cond = df['statename'] == state
    df[state_cond] = df[state_cond].ffill(axis=0)
    df[state_cond] = df[state_cond].bfill(axis=0)
  
  df = df.dropna()
  
  # Standardize:
  # to_scale = ['imports', 'exports']
  # scaler = StandardScaler()
  # df[to_scale] = scaler.fit_transform(df[to_scale])
  
  # Add surplus column
  df['surplus'] = df['exports'] - df['imports']
  
  # Add columns measuring change in exports/imports over one/five years
  
  df['expt_past_yr'] = df['exports'] - df['exports'].shift(periods=1)
  df['expt_past_5_yrs'] = df['exports'] - df['exports'].shift(periods=5)
  
  df['impt_past_yr'] = df['imports'] - df['imports'].shift(periods=1)
  df['impt_past_5_yrs'] = df['imports'] - df['imports'].shift(periods=5)
  df = df.bfill()
  return df

natl_trade = clean_natl_trade(natl_trade)
natl_trade
```

```
import pandas as pd
import numpy as np

dyadic_war_url = 'https://raw.githubusercontent.com/will-cotton4/COW-DataScience/master/directed_dyadic_war.csv'
national_trade_url = 'https://raw.githubusercontent.com/will-cotton4/COW-DataScience/master/National_COW_4.0.csv'
dyadic_trade_url = 'https://raw.githubusercontent.com/will-cotton4/COW-DataScience/master/Dyadic_COW_4.0.csv'
cc_url = 'https://raw.githubusercontent.com/will-cotton4/COW-DataScience/master/COW_country_codes.csv'

dyad_trade = pd.read_csv(dyadic_trade_url)
cc = pd.read_csv(cc_url)

def clean_dyad_trade(df):
  # This function cleans the dyadic trade dataset:
  # - Removes admin. and sparse variables (mentioned in natl_trade_clean)
  # - Removes years past 2007
  # - Imputes NaNs to missing values (-9)
  # - Adds column recording missing records
  # - Replaces missing values with forward/backfill for country
  # - Standardizes flow1, flow2, smoothflow1, 
  #   smoothflow2, smoothtotrade (commented out for now)
  # - TODO: Add imports to/exports from US, counter
  
  df = df.copy()
  
  df = df.drop(columns=['bel_lux_alt_flow1', 'bel_lux_alt_flow2',
                        'china_alt_flow1', 'china_alt_flow2', 
                        'source1', 'source2', 'version'])
  
#   flow1	flow2	smoothflow1	smoothflow2	smoothtotrade	
  df = df[df['year'] <= 2007]
  
  df = df.replace({-9:np.NaN})
  df['full_data'] = df[['flow1',	'flow2',	'smoothflow1',	'smoothflow2',	'smoothtotrade']].isnull().any(axis='columns')
  
  for country in df['importer1']:
    df[df['importer1']==country] = df[df['importer1']==country].fillna(method='ffill')
    df[df['importer1']==country] = df[df['importer1']==country].fillna(method='bfill')
    
  # Remaining values are from countries that totally lack data
  
  df = df.dropna()
  
  
  # Standardize:
  #   to_scale = ['flow1', 'flow2', 'smoothflow1', 'smoothflow2', 'smoothtotrade']
  #   scaler = StandardScaler()
  #   df[to_scale] = scaler.fit_transform(df[to_scale])
  
  return df

dyad_trade = clean_dyad_trade(dyad_trade)
dyad_trade

# print(dyad_trade.shape)
```
