## Data Cleaning Process

**NOTE**: the DataFrames used must be shaped into a (country, year, ...) format. For this reason I'm going to use the `dyad_yearly` DataFrame. It contains all important features (ccode, defense, neutrality, etc). Unlike `members_yearly` it also has a feature listing the allied state, which I predict will be a valuable feature for the machine learning model.

1. Drop `dyad_end_day` and `dyad_end_month`
    * These features are so sparsely populated that employing aggregate imputation methods will most likely only introduce noise to the data.
2. Drop `version4id` and `version`: unhelpful for prediction
3. Impute `nonaggression` and `entente`
    * Respectively, only 21 and 12 observations are missing.

For `entente` all missing obs. are for Iran from 1926-1937. During this time there were no wars involving Iran[1](https://en.wikipedia.org/wiki/List_of_wars_involving_Iran), though there were wars involving Turkey[2](https://en.wikipedia.org/wiki/List_of_wars_involving_Turkey). Due to the fact that their formal alliance only lasted until 1937, when Turkey took up arms against the Dersim Tribes, I think its likely that the alliance didn't involve an entente pact else Iran would've aided Turkey at this time.

For `nonaggression` all missing obs. are for UK with Russia, Germany, and Austria-Hungary. I couldn't find records online describing what kind of pact each country had with UK. Due to the short nature of the alliances I will impute with 0.

```python
# Make a deterministic cleaning function
def clean(df: pd.DataFrame):
    # Self-enforce immutability, "copy-on-write"
    X = df.copy()
    
    # Step 1
    X = X.drop(columns=['dyad_end_day', 'dyad_end_month'])
    
    # Step 2
    X = X.drop(columns=['version4id', 'version'])
    
    # Step 3
    X['entente'] = X['entente'].fillna(0)
    X['nonaggression'] = X['nonaggression'].fillna(0)
    
    return X
```

```python
cleaned_dyads = clean(dyads_yearly)
cleaned_dyads.info()
```
